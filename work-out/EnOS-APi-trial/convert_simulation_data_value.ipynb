```python
def transform_by_divide(csv_file, divider, remove_col ):
    data = pd.read_csv(csv_file)  
    
    # list(data) or 
    cols = list(data.columns)
    cols.remove(remove_col)
    print(cols)
    
    ## transform
    transformed_df = data.apply(lambda x: np.divide(x*, divider) if x.name in cols else x)
    return transformed_df;
```

```python
csv_file = '/Users/searoc/Documents/0-code/02-sgms/Test_automation/simulator/data/PPT4-6-PV-power-meter-July-Oct-2020.csv'
divider = -1000
remove_col = 'Timestamp'

df = transform_by_divide(csv_file, divider, remove_col)
df = df.set_index('Timestamp')
print(df)
df.to_csv(csv_file + ".transformed", encoding='utf-8', date_format='%Y-%m-%d MM-DD-HH')
```
