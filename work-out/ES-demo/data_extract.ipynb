Process the DAQ files from 03 Aug, and then save into new new csv files

```python
filePath = '/media/searoc/4T-backup/30-Envision/03 Aug 2020 DAQ Files/'
csvFiles = get_file_list(filePath, "zip", False)
fileExport = '/home/searoc/00-code/05-raw-data/narrow_down'

print('Read number of csv file: %d'% len(csvFiles))

wildcard = 'EdgeDetection'

for file_ in csvFiles:
    if (wildcard in file_) :
        continue
    try:
        df = load_zip_data(file_)
        fileName = os.path.basename(file_);
        dayFolder = getParent(file_, "/../../")

        craneFolder = getParent(file_, "/../../../")
        craneFolder = craneFolder.replace(dayFolder,'')

        if not (df is None):
            #print (df.head())
            targetFolder = fileExport + "/" + craneFolder + "/" + dayFolder
            if not os.path.exists(targetFolder):
                os.makedirs(targetFolder)

            export_dataset(df, fileName, targetFolder)
            print('exported: %s'% fileName)   
    except:
        print ('Caught ParserError in file: ' + file_)
    
```
